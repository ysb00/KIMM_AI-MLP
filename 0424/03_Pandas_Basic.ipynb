```python
# Example 1. Dataframe 생성
dict_data = {'c0':[1,2,3],'c1':[4,5,6],'c2':[7,8,9],'c3':[10,11,12],'c4':[13,14,15]}

df = pd.DataFrame(dict_data)

print(type(df))
print(df)
```

```python
# Example 2. Dataframe 이름 설정
df = pd.DataFrame([[15,'남','덕영중'],[17,'여','수리중']],
                 index = ['준서','예은'],
                 columns = ['나이','성별','학교'])

print(df)
print()
print(df.index)
print()
print(df.columns)
```

```python
# Example 3. 일부분의 행 인덱스와 열 이름 변경하기
df.rename(columns = {'나이':'연령','성별':'남녀','학교':'소속'}, inplace = True)

df.rename(index = {'준서':'학생1','예은':'학생2'}, inplace = True)

print(df)
```

```python
# Example 4. 행 삭제
exam_data = {'수학' : [90,80,70],'영어':[98,89,95],'음악':[85,95,100],'체육':[100,90,90]}

df = pd.DataFrame(exam_data, index = ['서준','우현','인아'])
print(df)
print()

df2 = df[:]
df2.drop('우현', inplace = True)
print(df2)
print()

df3 = df[:]
df3.drop(['우현','인아'], axis=0, inplace = True)
print(df3)
print()

df4 = df[:]
df4.drop(['우현','인아'], axis=0, inplace = False)
print(df4)
```

```python
# Example 5. 열 삭제
df4 = df[:]
df4.drop('수학', axis = 1, inplace = True)
print(df4)
print()

df5= df[:]
df5.drop(['영어','음악'], axis=1, inplace=True)
print(df5)
```

```python
# Example 6. 행 선택 loc 혹은 iloc
label1 = df.loc['서준']
position1 = df.iloc[0]
print(label1)
print()
print(position1)
```

```python
# Example 7. 열 선택 loc/iloc 혹은 직접 필터링
label2 = df.loc[['서준','우현']]
print(label2)

label3 = df[['음악','체육']]
print(label3)
```

```python
# Example 8. 열 추가
df['국어'] = 80
df
```

```python
# Example 9. 행 추가
df.loc[3]=0
df
```

```python
# Example 10. 행 추가
df.loc[4] = ['동규',90,80,70,60]
df
```

```python
# Example 11. 행, 열의 위치 바꾸기
df = pd.DataFrame(exam_data, index = ['서준','우현','인아'])
print(df)
print()

df=df.transpose()
print(df)
```

```python
# Example 12. 행번호(index) 초기화
df=df.transpose().reset_index()
df
```

```python
# Example 13. 특정 열을 행번호로 설정
df = df.set_index(['index'])
df
```

```python
# Example 14. c1 열을 기준으로 내림차순 정렬
df = df.sort_values(by='음악',ascending = False)
df
```

```python
# Example 15. 열 순서 변경
import seaborn as sns
titanic=sns.load_dataset('titanic')
df = titanic.loc[0:4,'survived':'age']
df
```

```python
# Example 16. List 생성 및 열 순서 재설정
columns_customed = ['pclass','sex','age','survived']
df_customed = df[columns_customed]
df_customed
```

```python
# Example 17. 리스트를 데이터프레임으로 변환
new_list = [2100, 1600, 1800, 1000, 1400]
new_df = pd.DataFrame(new_list,columns=['income'])
new_df
```

```python
# Example 18. 데이터프레임 병합 (열 기준)
df_customed = pd.concat([df_customed, new_df], axis=1)
df_customed
```

```python
# Example 19. 데이터프레임 병합 (행 기준)
df_customed = pd.concat([df_customed.iloc[0:2,:], df_customed.iloc[1:3,:]], axis=0)
df_customed
```

```python
# Example 23. 그룹 객체 만들기
titanic_groupby = titanic.groupby('pclass').count()
titanic_groupby
```

```python
# Example 23. 그룹 객체 만들기 및 색 표현
titanic_groupby = titanic.groupby('pclass')[['sex','age']].count().style.background_gradient(cmap='Purples')
titanic_groupby
```

```python
# Example 23. 그룸 내 통계량 연산
titanic.groupby('pclass')['age'].agg([max,min])
```

```python
# Example 23. 그룸 내 통계량 연산 (컬럼별로 서로 다른 통계량 설정)
agg_format = {"age":"max","sibsp":"sum","fare":"mean"}
titanic.groupby('pclass').agg(agg_format)
```
