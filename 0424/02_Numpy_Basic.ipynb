```python
# Example 1. 1D array
x = np.array([1,2,3])
print(x)
```

```python
# Example 2. 2D array
x = np.array([[1,2,3,],[4,5,6]])
print(x)
```

```python
# Example 3. 순차형 1D array
np.arange(10)
```

```python
# Example 4. 일정 간격의 1D array
np.arange(4,10,2)
```

```python
# Example 5. 일정 간격의 20등분한 1D array
np.linspace(1,10,20)
```

```python
# Example 6. 1로 채워진 array
np.ones((3,3))
```

```python
# Example 7. 0으로 채워진 array
np.zeros((2,4,3)) # (page, row, column)
```

```python
# Example 8. 0과 1 사이의 균일 분포 난수 생성
np.random.rand(5)
```

```python
# Example 9. 0과 1 사이의 균일 분포 난수 생성(2D array)
np.random.rand(2,3)
```

```python
# Example 10. 가우시안 표준 정규 분포 난수 생성
np.random.randn(5)
```

```python
# Example 11. 난수값 고정
np.random.seed(0)
np.random.rand(5)
```

```python
# Example 12. 랜덤 정수 생성
arr = np.random.randint(10,20,size=(2,3,3))
print(arr)

print("\n 첫번째 arr:"); print(arr[0])

print("\n 두번째 arr:"); print(arr[1])
```

```python
# Example 13. 특정 위치의 값 변경
arr[1,1,2] = 120
arr
```

```python
# Example 14. array 원소 필터링
np.random.seed(2)
x = np.random.randint(1,10,8)
print("arr:"); print(x)
print("\nFiltered arr:"); print(x[x>3])
```

```python
# Example 15. 조건에 맞는 값의 index 반환
np.where(x>3)
```

```python
# Example 16. Array 형태(차원) 변경
b = np.arange(15)
b
```

```python
b = np.arange(15).reshape(3,5)
b
```

```python
np.ravel(b) # 원본 그대로 사용
```

```python
b.flatten() # copy 사용
```
